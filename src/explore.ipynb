```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/linear-regression-project-tutorial/main/medical_insurance_cost.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.info()
```

```python
#factorizamos categoricas 


for columna in df.columns:
    if df[columna].dtype  not in ('int64','float64'):
        df[columna]= pd.factorize(df[columna])[0]
df.info()
```

```python
for columna in df.columns:
    if columna not in ('charges'):
        fig, axis=plt.subplots(2,1,figsize =(10,7))
        sns.regplot(df,x=columna,y='charges',ax=axis[0])
        sns.heatmap(df[[columna,'charges']].corr(),annot=True,fmt=".2f",ax=axis[1],cbar=False)
        plt.tight_layout()
        plt.show()
```

```python
x=df[['smoker','age','bmi']]
y=df['charges']

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
X_train.head()

```

```python
for columna in X_train.columns:
    sns.histplot(X_train,x=columna)
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
#graficos univariables me indican si usar MinMaxScaler, si la mayoria asemejan la distribucion normal tendria que usar un standar scaler
scaler = MinMaxScaler()

X_train_esc = scaler.fit_transform(X_train,y_train)
X_test_esc = scaler.transform(X_test)

X_train_esc= pd.DataFrame(X_train_esc,columns=x.columns,index=X_train.index)
X_test_esc= pd.DataFrame(X_test_esc,columns=x.columns,index=X_test.index)

X_train_esc.head()
```

```python
from sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(X_train_esc,y_train)

print(f"intercepto (a):{model.intercept_}")
print(f"coeficientes (b1 , b2 , b3):{model.coef_} ")
```

```python
y_pred = model.predict(X_test_esc)
y_pred
```

```python
from sklearn.metrics import root_mean_squared_error,r2_score

print(f"Error cuadratico medio en raiz (RMSE): {root_mean_squared_error(y_test,y_pred)}")
print(f"R2: {r2_score(y_test,y_pred)}")
```
